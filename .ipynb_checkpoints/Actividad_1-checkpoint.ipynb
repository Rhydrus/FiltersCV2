## 1.2 **Agregar ruido Sal y Pimienta**

```python
def ruido(imagen1, probabilidad):
    filas, columnas, canales = imagen1.shape
    ruido = np.zeros(imagen1.shape, np.uint8)
    
    for fila in range(filas):
        for columna in range(columnas):
            Random = random.random()
            if Random < probabilidad:
                ruido[fila][columna][:] = 0
            elif Random > (1 - probabilidad):
                ruido[fila][columna][:] = 255
            else:
                ruido[fila][columna][:] = imagen1[fila][columna][:]
    return ruido

imagenRuido1 = ruido(imagen01, 0.05)
plt.imshow(imagenRuido1)
plt.title("Imagen con Ruido de Sal y Pimienta")
plt.show()
"""


imagenRuido2 = ruido(imagen02, 0.05)
# Mostrar las imágenes


"""
```

### **1.3.2 Filtro Propio - Filtro de media**

```python
# Matriz kernel
def filtro(imagen_ruido, tamano_kernel):
    filas, columnas, canales = imagen_ruido.shape
    imagen_array = np.array(imagen_ruido)
    resultado = np.zeros(imagen_ruido.shape, np.uint8)
    imagen_extendida = np.pad(imagen_array, ((tamano_kernel // 2, tamano_kernel // 2), (tamano_kernel // 2, tamano_kernel // 2), (0, 0)), mode='reflect')

    for i in range(filas):
        for j in range(columnas):
            # Extraer la región del kernel para cada canal
            vecindario = imagen_extendida[i:i + tamano_kernel, j:j + tamano_kernel, :]

            # Calcular la mediana por canal y asignar al píxel correspondiente en la imagen de resultado
            resultado[i, j, :] = np.median(vecindario, axis=(0, 1))

    return resultado
    
resultado_filtro = filtro(imagenRuido1, tamano_kernel=3)
plt.imshow(resultado_filtro)
plt.title("Filtro")
plt.show()
```
